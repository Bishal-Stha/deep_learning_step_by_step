```python
import torch
import torch.nn as nn

# 1. THE DATA (Inputs: Hours studied | Outputs: 0=Fail, 1=Pass)
X = torch.tensor([[2.0], [3.0], [7.0], [8.0]], dtype=torch.float32)
y = torch.tensor([[0.0], [0.0], [1.0], [1.0]], dtype=torch.float32)

# 2. THE ARCHITECTURE (A single linear layer + Sigmoid to get a probability)
class SimpleClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer = nn.Linear(1, 1) # Takes 1 input number, outputs 1 number
        self.sigmoid = nn.Sigmoid()   # Squeezes the output between 0 and 1
        
    def forward(self, x):
        return self.sigmoid(self.layer(x))

# Initialize our model
model = SimpleClassifier()

# 3. THE LOSS & OPTIMIZER
criterion = nn.BCELoss() # Binary Cross Entropy (Standard for 0 or 1 problems)
optimizer = torch.optim.SGD(model.parameters(), lr=0.1) # Simple Gradient Descent

# 4 & 5. THE TRAINING LOOP (Let's run it for 100 turns)
print("🏋️ Training the network...")
for epoch in range(100):
    # Forward Pass: Get the model's current guess
    predictions = model(X)
    
    # Calculate how bad the guess was
    loss = criterion(predictions, y)
    
    # Reset old calculations, calculate new directions, and take a step
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1}/100 | Loss: {loss.item():.4f}")

print("\n✅ Training Complete!")

# TEST THE MODEL
with torch.no_grad(): # Turn off learning mechanics for testing
    test_student = torch.tensor([[9.0]]) # A student who studied 9 hours
    prediction = model(test_student)
    print(f"🔮 Probability of a student passing who studied 9 hours: {prediction.item()*100:.2f}%")
```
